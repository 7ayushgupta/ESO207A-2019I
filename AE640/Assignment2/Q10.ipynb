```python
import matplotlib.pyplot as plt
import scipy
import numpy as np
```

```python
def solve_covariance_matrix(covariance_matrix):
    a = covariance_matrix[0][0]
    b = covariance_matrix[0][1]
    c = covariance_matrix[1][1]
    
    lambda1 = (a+c)/2 + np.sqrt( ((a-c)/2)**2 + b**2 )
    lambda2 = (a+c)/2 - np.sqrt( ((a-c)/2)**2 + b**2 )
    
    if(b==0 and a>=c):    theta = 0
    elif(b==0 and a<c):   theta = np.pi/2
    else:                 theta = np.arctan2(lambda1 - a, b)
    return lambda1, lambda2, theta
```

## Subpart 1 - mean and covariance

```python
sigma_theta = (np.pi/180.0)*10
sigma_r = 0.01
```

```python
mean_x = 0 
mean_y = np.exp(-sigma_theta**2/2)
print(mean_x, mean_y)
```

```python
var_x = (1 + sigma_r**2)*(0.5 * (1 - np.exp(-2*sigma_theta**2)))
var_y = (1 + sigma_r**2)*(0.5 * (1 + np.exp(-2*sigma_theta**2)))
cov_xy = 0
covariance_matrix = np.array([[var_x, cov_xy],
                             [cov_xy, var_y]])
print(covariance_matrix)
```

```python
lambda1, lambda2, theta_ellipse = solve_covariance_matrix(covariance_matrix)
r = 1 + np.random.normal(0, sigma_r, 1000)
theta = np.pi/2 + np.random.normal(0, sigma_theta, 1000)
x = r*np.cos(theta)
y = r*np.sin(theta)
```

```python
from matplotlib.patches import Ellipse
plt.figure(figsize=(10,8))
ax = plt.gca()
plt.scatter(x, y, s=0.1)
ell = Ellipse(xy=(mean_x, mean_y),
                  width=np.sqrt(lambda1), height=np.sqrt(lambda2),
                  angle=theta_ellipse, edgecolor='r', fc='None', lw=1)
ax.add_artist(ell)
ax.set_xlim(-1, +1)
ax.set_ylim(0.7, +1.2)
plt.title("Mean and Covariance Ellipse")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.show()
```
