```python
# import prutorlib
import matplotlib
import matplotlib.pyplot as plt
from pylab import *
########## code here

alpha = 1
xi = 0
xf = 5

dt = 0.005
dx = 0.01

nx = int((xf-xi)/dx)
nt = int(3/dt)

x = np.arange (xi-2*dx, xf+3*dx, dx)
y = np.zeros((nt+1,len(x)))
y[0] = np.exp(-32*(x-4)**2)


for i in range (nt):
    for j in range (2, len(x)-3):
        k1 = -alpha*dt*(y[i][j]-y[i][j-2])/(2*dx)
        a1 = y[i][j-1] + k1/2 
        k2 = -alpha*dt*(y[i][j+1]-y[i][j-1])/(2*dx)
        a2 = y[i][j] + k2/2
        k3 = -alpha*dt*(a2-a1)/(2*dx)
        y[i+1][j] = y[i][j]+k3
        
for i in range(0,nt + 1, nt//3):
    plt.plot(x[2:len(x)-2],y[i][2:len(x)-2])
plt.xlabel('x values')
plt.ylabel('Psi values')
plt.title( '1D wave equation')
plt.gca().legend(('t=0.0','t=1','t=2','t=3'))
# prutorlib.plot(plt, "output.png")
```
